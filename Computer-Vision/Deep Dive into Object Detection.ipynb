### Task 1: Intersection over Union (IoU)
Object detectors often predict multiple bounding boxes for the exact same object. To clean this up, the Non-Maximum Suppression (NMS) algorithm deletes duplicate boxes.

To know if two boxes are duplicates, NMS relies on IoU.

Your task is to complete the calculate_iou function.

* A box is defined as a list: [x_min, y_min, x_max, y_max]
* Hint 1: The width of a box is (x_max - x_min).
* Hint 2: The intersection rectangle's x_min is the maximum of the two boxes' x_mins.

```python
def calculate_iou(box1, box2):
    """
    Calculate the Intersection over Union (IoU) of two bounding boxes.
    Box format: [x_min, y_min, x_max, y_max]
    """

    # 1. Calculate the coordinates of the intersection rectangle
    # TODO: Find x_left, y_top, x_right, y_bottom of the overlapping area

    x_left = max(box1[0], box2[0])
    y_top = max(box1[1], box2[1])
    x_right = min(box1[2], box2[2])
    y_bottom = min(box1[3], box2[3])


    # 2. Check if there is an intersection at all!
    # If x_right < x_left or y_bottom < y_top, they don't overlap. Return 0.0
    # TODO: Add the check here
    if x_right < x_left or y_bottom < y_top:
        return 0.0
    # 3. Calculate the area of intersection
    # TODO: intersection_area = ...
    intersection_area = (x_right - x_left) * (y_bottom - y_top)
    # 4. Calculate the area of both individual bounding boxes
    box1_area = (box1[2] - box1[0]) * (box1[3] - box1[1])
    # TODO: box2_area = ...
    box2_area =( box2[2] - box2[0]) * (box2[3] - box2[1])


    # 5. Calculate the area of the Union
    # Union Area = Area 1 + Area 2 - Intersection Area
    # TODO: union_area = ...
    union_area = box1_area + box2_area - intersection_area
    # 6. Compute IoU
    # TODO: iou = intersection_area / union_area
    # return iou
    eps=1e-6
    iou = intersection_area / (union_area+eps)
    return iou


# --- TEST YOUR CODE ---
box_A = [50, 50, 150, 150]
box_B = [100, 100, 200, 200]
box_C = [200, 200, 300, 300]

iou_AB = calculate_iou(box_A, box_B)
iou_AC = calculate_iou(box_A, box_C)

print(f"IoU between Box A and Box B (Should be ~0.142): {iou_AB:.3f}")
print(f"IoU between Box A and Box C (Should be 0.000): {iou_AC:.3f}")
```
